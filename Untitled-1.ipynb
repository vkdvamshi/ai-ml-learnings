```python
#!/usr/bin/env python3
# Script to remove orphan data security context
# /fscmRestApi/resources/11.13.18.05/dataSecurities?finder=ManageDataAccessRF;Userrf=ckiernat@watlow.com

import pandas as pd
import numpy as np
import json
from tenacity import retry
import requests

pd.set_option('display.width', 1000)
pd.set_option('display.max_columns', 1000)
headers = {'Content-Type': 'application/json'}
urls = {"dev1": "https://fa-ewrc-dev1-saasfaprod1.fa.ocs.oraclecloud.com" }
targetEnv = urls['dev1']  # set this for each env
apis = {"users": "/hcmRestApi/scim/Users",
        "updateRole": "/hcmRestApi/scim/Roles/{id}",
        "bulkOp": "/hcmRestApi/scim/Bulk",
        "roles": "/hcmRestApi/scim/Roles",
        "dataSecurities": """/fscmRestApi/resources/11.13.18.05/dataSecurities""",
        "userAccts" : "/hcmRestApi/resources/11.13.18.05/userAccounts?limit=500" }
creds = {"userName": 'ShobanaDevi.Ts@watlow.com',
         "Password": 'Welcome123'}

dataSecurityPayLoad = """{
"SecurityContext" : "Ledger",
"SecurityContextValue" : "BalTranSCb=CALXCUR",
"RoleNameCr" : "Financial Analyst",
"UserName" : "FUSION_APPS_FSCM_ENT_MANAGER"
}"""

dataSecurityPayLoad = json.loads(dataSecurityPayLoad)

@retry()
def fetch_all_user_accounts(base_url, username, password):
    # Initial URL for the userAccounts resource
    # The 'limit' parameter is specified to control page size
    url = f"{base_url}/hcmRestApi/resources/11.13.18.05/userAccounts?limit=500"
    all_accounts = []
    while url:
        try:
            # Make the GET request with basic authentication
            response = requests.get(url, auth=(username, password), headers={'Accept': 'application/json'})
            response.raise_for_status() # Raise an exception for bad status codes (4xx or 5xx)
            data = response.json()

            # The actual records are typically within a 'items' or similar key in the JSON response
            if 'items' in data:
                all_accounts.extend(data['items'])
            elif 'Resources' in data: # SCIM endpoints might use 'Resources'
                all_accounts.extend(data['Resources'])
            
            # Check for a 'next' link in the response to determine the next page URL
            # Oracle APIs often provide HATEOAS links in the response metadata
            url = None
            if 'links' in data:
                for link in data['links']:
                    if link.get('rel') == 'next':
                        url = link.get('href')
                        break
            
            # For APIs that use offset/limit parameters directly, you may need to manually
            # construct the next URL using 'offset' and 'limit' values found in the response metadata
            # e.g., if 'offset' and 'limit' keys are present in the response
            # next_offset = data.get('offset', 0) + data.get('limit', 25)
            # total_results = data.get('totalResults', 0)
            # if next_offset < total_results:
            #     url = f"{base_url}/hcmRestApi/resources/11.13.18.05/userAccounts?limit=25&offset={next_offset}"


        except requests.exceptions.RequestException as e:
            print(f"Error fetching data: {e}")
            break
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
            break

    return all_accounts


def fetch_all_users():
    result = []
    url = targetEnv + apis['users'] +"?startIndex={}&count=500"
    totalResults = requests.get(url.format(1), headers=headers, auth=(creds['userName'], creds["Password"])).json().get('totalResults', 0)
    for i in range(1, totalResults, 500):
        try:
            response = requests.get(url.format(i), headers=headers, auth=(creds['userName'], creds["Password"]))
            if response.status_code == 200:
                users = response.json().get('Resources', [])
                if users is not None:
                    result.extend(users)
            else:
                print(f"Failed to fetch users at index {i}: {response.status_code} - {response.text}")
        except Exception as e:
            print(f"Exception while fetching users at index {i}: {e}")
    return pd.json_normalize(result) # returns json of all users.

def fetch_all_roles_usr():
    result = []
    url = targetEnv + apis['roles'] +"?startIndex={}&count=500"
    totalResults = requests.get(url.format(1), headers=headers, auth=(creds['userName'], creds["Password"])).json().get('totalResults', 0)
    for i in range(1, totalResults, 500):
        try:
            response = requests.get(url.format(i), headers=headers, auth=(creds['userName'], creds["Password"]))
            if response.status_code == 200:
                roles = response.json().get('Resources', [])
                if roles is not None:
                    result.extend(roles)
            else:
                print(f"Failed to fetch roles at index {i}: {response.status_code} - {response.text}")
        except Exception as e:
            print(f"Exception while fetching roles at index {i}: {e}")
    return result # returns json of all roles.

def fetch_all_data_securities(usrName):
    result = []
    url = targetEnv + apis['dataSecurities'] + """?finder=ManageDataAccessRF;Userrf="{}"&limit=500""".format(usrName)
    try:
        print ( url )
        response = requests.get(url, headers=headers, auth=(creds['userName'], creds["Password"]))
        if response.status_code == 200:
            dataSecurities = response.json().get('items', [])
            if dataSecurities is not None:
                return pd.json_normalize(dataSecurities) # returns json of all data securities for the user.
        else:
            print(f"Failed to fetch data securities at index: {response.status_code} - {response.text}")
    except Exception as e:
        print(f"Exception while fetching data securities at index: {e}")
```
